```python
import hail as hl
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
%ls "UKB_DATA/tsv_pheno/"
```

```python
DATA_PATH = "UKB_DATA/tsv_pheno"
df0=pd.read_csv(f"{DATA_PATH}/lv_ef_UKB.tsv",sep='\t')
df2=pd.read_csv(f"{DATA_PATH}/lv_ef_n_sd2.tsv",sep='\t')
df4=pd.read_csv(f"{DATA_PATH}/lv_ef_n_sd4.tsv",sep='\t')
df6=pd.read_csv(f"{DATA_PATH}/lv_ef_n_sd6.tsv",sep='\t')
df8=pd.read_csv(f"{DATA_PATH}/lv_ef_n_sd8.tsv",sep='\t')
df10=pd.read_csv(f"{DATA_PATH}/lv_ef_n_sd10.tsv",sep='\t')
```

```python
colors= ["#41354d", "#78416d", "#b53b73","#d23d69", "#e67d62","#f7b795"]
fig = plt.figure(figsize=(8,8)) # default is (8,6)
plt.scatter(df0['pheno'], df10['pheno'],c=colors[0],alpha=0.8,label="MAE=7.92")
plt.scatter(df0['pheno'], df8['pheno'],c=colors[1],alpha=0.8, label="MAE=6.38")
plt.scatter(df0['pheno'], df6['pheno'],c=colors[2],alpha=0.8, label="MAE=4.79")
plt.scatter(df0['pheno'], df4['pheno'],c=colors[3],alpha=0.8, label="MAE=3.18")
plt.scatter(df0['pheno'], df2['pheno'],c=colors[4],alpha=0.8, label="MAE=1.59")
plt.legend()
plt.xlabel("Base LVEF")
plt.ylabel("LVEF with Noise")
plt.xlim(-20,110)
plt.ylim(-20,110)
dat = np.linspace(0,100)
plt.plot(dat,dat, 'w--')
plt.tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)
plt.savefig("overlap_scatter.png",dpi=600)
```
