```python
rsid = []
betahat = []
se = []
chr = []
pos = []
for i in range(0,21):
    rsid.append(results[i]['rsid'])
    betahat.append(results[i]['beta'])
    se.append(results[i]['standard_error'])
    chr.append(results[i]['chr'])
    pos.append(results[i]['pos'])
```

```python
for i in range(1,21):
    rsid[0]=[y for x in [rsid[0], rsid[i]] for y in x]
    betahat[0]=[y for x in [betahat[0], betahat[i]] for y in x]
    se[0]=[y for x in [se[0], se[i]] for y in x]
    chr[0]=[y for x in [chr[0], chr[i]] for y in x]
    pos[0]=[y for x in [pos[0], pos[i]] for y in x]
```
